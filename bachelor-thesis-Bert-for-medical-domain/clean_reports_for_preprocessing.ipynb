```python
text_to_id = {}
for file in files:
    with open(file, mode='r', encoding="latin-1") as f:
        title = file[-16:]
        text_to_id[title] = f.read()
```

```python
check_texts = list(text_to_id.values())
```

```python
checks_cp = list(set(check_texts)).copy()
ids = []
for k,v in text_to_id.items():
    if( k not in ids ) & ( v in checks_cp):
        ids.append(k)
        checks_cp.remove(v)
```
